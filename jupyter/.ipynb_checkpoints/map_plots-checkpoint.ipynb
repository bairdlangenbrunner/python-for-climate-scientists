```python
import matplotlib.pyplot as mp
import cartopy
```

# Homolosine for _projects_ page

```python
fig, ax = mp.subplots(1,1,subplot_kw={'projection':cartopy.crs.InterruptedGoodeHomolosine()})
fig.set_size_inches(12,6)

ax.coastlines(resolution='110m', lw=1.5, color='#002b36')
ax.outline_patch.set_edgecolor('#002b36')
ax.outline_patch.set_linewidth(1.5)

#ax.outline_patch.set_visible(False)
ax.background_patch.set_visible(False)

fig.tight_layout()
fig.savefig('homolosine_map.svg', transparent=True, bbox_inches='tight')
```

# Robinson for *pforcs* homepage

```python
map_proj = cartopy.crs.Robinson(central_longitude=12)
geodetic_proj = cartopy.crs.Geodetic()

fig, ax = mp.subplots(1,1,subplot_kw={'projection':map_proj})
fig.set_size_inches(12,6)

#ax.add_feature(cartopy.feature.LAND, zorder=0, color='#002b36')
#ax.coastlines(resolution='110m', lw=1.5, color='#002b36')
#ax.outline_patch.set_edgecolor('#002b36')

ax.add_feature(cartopy.feature.LAND, zorder=0, color='#fdf6e3')
ax.coastlines(resolution='110m', lw=1.5, color='#fdf6e3')
ax.outline_patch.set_edgecolor('#fdf6e3')

ax.outline_patch.set_linewidth(1.5)

x0,x1,y0,y1 = ax.get_extent()
ax.set_extent([x0,x1,0.7*y0,y1], crs=map_proj)

ax.outline_patch.set_visible(False)
ax.background_patch.set_visible(False)

fig.tight_layout()
fig.savefig('robinson_map.svg', transparent=True, bbox_inches='tight')
```
