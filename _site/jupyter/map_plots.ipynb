```python
import matplotlib.pyplot as mp
import cartopy
```

# Homolosine for _projects_ page

```python
fig, ax = mp.subplots(1,1,subplot_kw={'projection':cartopy.crs.InterruptedGoodeHomolosine()})
fig.set_size_inches(13,13)

ax.coastlines(resolution='110m', lw=1.5, color='#002b36')
ax.outline_patch.set_edgecolor('#002b36')
ax.outline_patch.set_linewidth(1.5)

#ax.outline_patch.set_visible(False)
ax.background_patch.set_visible(False)

fig.tight_layout()
fig.savefig('homolosine_map.svg', transparent=True, bbox_inches='tight')
```

# Robinson

```python
fig, ax = mp.subplots(1,1,subplot_kw={'projection':cartopy.crs.Robinson(central_longitude=12)})
fig.set_size_inches(13,13)

ax.add_feature(cartopy.feature.LAND, zorder=0, color='#002b36')

ax.coastlines(resolution='110m', lw=1.5, color='#002b36')
ax.outline_patch.set_edgecolor('#002b36')
ax.outline_patch.set_linewidth(1.5)

#ax.set_extent([-8625155.12857459, 8625155.12857459,-17005833.33052523, 17005833.33052523])

ax.outline_patch.set_visible(False)
ax.background_patch.set_visible(False)

fig.tight_layout()
#fig.savefig('homolosine_map.svg', transparent=True, bbox_inches='tight')
```
